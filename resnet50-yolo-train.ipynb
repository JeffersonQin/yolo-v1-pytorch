```python
%matplotlib inline
%config InlineBackend.figure_formats = ['svg']
from utils import load_data_voc
from yolo import *
from torch import nn
import torchvision
```

```python
train_iter, test_iter, test_iter_raw = load_data_voc(batch_size=4)
```

```python
resnet50 = torchvision.models.resnet50(pretrained=True)
backbone = nn.Sequential(*list(resnet50.children())[:-2]) # remove avg pool and fc
net = Yolo(backbone, backbone_out_channels=2048)
```

```python
def lr(epoch):
    if epoch < 10: return 0.001 * (epoch + 1)
    if epoch < 85: return 0.01
    if epoch < 115: return 0.001
    if epoch < 145: return 0.0001

train(net, train_iter, test_iter, 145, lr=lr, momentum=0.9, weight_decay=5e-4, accum_batch_num=16, save_path='./model', pretrained=True)
```
